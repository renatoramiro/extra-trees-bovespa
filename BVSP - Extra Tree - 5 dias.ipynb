```python
import pandas as pd
import numpy as np
```

## Carregando os dados

```python
df = pd.read_csv('best_features_bvsp_close_classifier_5_days.csv')
df['Date'] = pd.to_datetime(df['Date'], )
df.set_index('Date', drop=True, inplace=True)
df
```

## Divisão de dados de treinamento e teste: Train Test Split

```python
X = df.drop(columns=['Target', 'Returns'], axis=1)
y = df['Target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                               test_size= 0.2 ,shuffle=False)
```

## Scaling

```python
from sklearn.preprocessing import QuantileTransformer

scaler = QuantileTransformer()

X2_train = X_train.values
y2_train = y_train.values

scaler.fit(X2_train)

X2_train = scaler.transform(X2_train)

X2_test = scaler.transform(X_test.values)
y2_test = y_test.values
```

## Construindo o modelo preditivo

```python
from sklearn.ensemble import ExtraTreesClassifier

model = ExtraTreesClassifier(n_estimators=17, max_features='sqrt',
                             max_depth=5, min_samples_split=2,
                             min_weight_fraction_leaf=0.0, 
                             n_jobs=-1, random_state=122)

model.fit(X2_train, y2_train)
```

### Score do Treinamento

```python
print(model.score(X2_train, y2_train))
```

### Realizando as predições

```python
predictions = model.predict(X2_test)
```

## Avaliando os resultados

### Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(y2_test, predictions))
```

### Cálculo da Curva ROC

Captura de tela 2020-12-07 145055.png

```python
import matplotlib.pyplot as plt
from sklearn.metrics import plot_roc_curve
ax = plt.gca()
model_auc = plot_roc_curve(model, X2_test, y2_test, ax=ax)
plt.show()
```

## Exibindo as probabilidades

```python
predictions_prob = model.predict_proba(X2_test)
predictions_prob
```

```python
predictions_prob = pd.DataFrame(predictions_prob)
predictions_prob.index = y_test.index

predictions = pd.DataFrame(predictions)
predictions.index = y_test.index

Tabela=pd.concat([predictions_prob,predictions,y_test],axis=1)
Tabela.dropna(inplace=True)

Tabela.columns=['Prob_de_ser_0','Prob_de_ser_1','Pred','Y']
Tabela.tail(5)
```

### Score de Treinamento e Teste

```python
print('Score Extra Trees Classifier no treinamento', model.score(X2_train, y2_train))
print('Score Extra Trees Classifier no teste', model.score(X2_test, y2_test))
```
